```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("../data/raw/road_trip_data.csv")

# View data
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Distance vs Travel Time
plt.figure()
sns.scatterplot(x="distance_km", y="travel_time_hr", data=df)
plt.title("Distance vs Travel Time")
plt.show()
```

```python
# Traffic Level Count
plt.figure()
sns.countplot(x="traffic_level", data=df)
plt.title("Traffic Level Distribution")
plt.show()
```
