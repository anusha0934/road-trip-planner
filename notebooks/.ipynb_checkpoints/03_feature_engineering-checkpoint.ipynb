```python
import pandas as pd

# Load raw data
df = pd.read_csv("../data/raw/road_trip_data.csv")

# Create new features
df["avg_speed"] = df["distance_km"] / df["travel_time_hr"]
df["cost_per_km"] = df["fuel_cost"] / df["distance_km"]

# View new features
df.head()
```

```python
df.to_csv("../data/processed/road_trip_features.csv", index=False)
print("Feature engineering completed")
```
